###   好用的Python可视化利器Matplotlib


首先需要对Matplotlib有一个了解

- matplotlib是什么？

    matplotlib是一个Python的绘图库，他以各种硬拷贝格式和跨平台的交互式环境生成初版质量级的图形
    
- 为什么选择Matplotlib?
    
    如果某天你发现自己要学习 Matplotlib, 很可能是因为: 
    
    1、Matplotlib 是一个非常强大的 Python 画图工具；
    
    2、手中有很多数据, 可是不知道该怎么呈现这些数据。
    
- Matplotlib 能帮你做什么？

    绘制线图、散点图、等高线图、条形图、柱状图、3D 图形、、甚至是图形动画等等。 
    
    优点：本身matplotlib非常强大，是一个高定制的内容，比如里面的颜色，线条粗细、形状等
    
    缺点：API太多，根本记不住，其实大家不需要全部记这些内容，
    
    只需要记住常见高频使用的一些函数或者属性就可以，后期遇到复杂图形或者不常见的图形的时候可以查阅相关资料，

### 网格线

```python
plt.grid(True)
```

```python
#定于一个x,其中np.pi表示180度，点三个参数0.01表示间隔
x = np.arange(-np.pi,np.pi,0.01)

plt.plot(x,np.sin(x),x,np.cos(x))
#想将上面的正玄和余弦线放置网格中，直接调用plt模块中的grid函数
plt.grid()
```

### 修改线条的样式

- linestyle：线的风格类型
- color：颜色
- marker:标记

```python
#先生成几个数据
x1 = np.random.randn(100)
x2 = np.random.randn(100)
x3 = np.random.randn(100)

#求上面x1,x2,x3的累加和，根据这个3个值绘制图形
plt.plot(x1.cumsum(), color='blue', linestyle='--', marker='>')
plt.plot(x2.cumsum())
plt.plot(x3.cumsum())
#在绘制下图例
plt.legend(['x1','x2','x3'])
```

### 保存图片
- savefig函数

```python
x1 = np.random.randn(100)
x2 = np.random.randn(100)
x3 = np.random.randn(100)

plt.plot(x1.cumsum(),color='blue',linestyle = '--',marker = 'o')
plt.plot(x2.cumsum(),color='green',linestyle = '-.',marker = '>')
plt.plot(x3.cumsum(),color='black',linestyle = ':',marker = 'D')
#在绘制下图例
plt.legend(['x1','x2','x3'])

#保存这张图片
plt.savefig('pic4.jpg',dpi=150,facecolor='green')
```

### x，y轴坐标刻度

- xticks:源码解说：获取或设置X轴的当前刻度位置和标签。
- yticks：源码解说：获取或设置Y轴的当前刻度位置和标签。

#### 通过xticks和yticks方式

```python
x = np.random.randn(100)
plt.plot(x.cumsum())
#绘制坐标
plt.xticks(np.linspace(0,100,7),list('ABCDEFG'))
#设置y轴的标签并设置大小
plt.yticks(np.linspace(-10,20,3),['min',0,'max'],fontsize = 15)
#还可以对x轴设置标签的角度
plt.xticks(np.linspace(0,100,7),list('ABCDEFG'),rotation = 90)
```

如何实现?

下面是一个包含cos、sin、sqrt函数的完整图像：
![](../集训营课件/img.png)

```python
%matplotlib inline
```

```python
x = np.arange(0,10,0.2)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.sqrt(x)

plt.figure(figsize=[20,9], dpi=90)
plt.plot(x, y1,color = 'blue', linewidth=1.5, linestyle='-', marker='.', label=r'$y = cos{x}$')
plt.plot(x, y2,color = 'green', linewidth=1.5, linestyle='-', marker='*', label=r'$y = sin{x}$')
plt.plot(x, y3,color = 'r', linewidth=1.5, linestyle='-', marker='x', label=r'$y = \sqrt{x}$')

#坐标轴的上下移动
ax = plt.subplot(111)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#移动边框线
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

#设置x y 取值范围
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(-1.5, 4.0)

#设置x y轴的刻度值
plt.xticks([2,4,6,8,10])
plt.yticks([-1.0, 0.0, 1.0, 2.0, 3.0,4.0])
plt.text(0.7,0.7,'h',color='g',fontsize=15)
plt.text(0.7,0.9,'t',color='g',fontsize=15)

#设置标题
plt.title('the function of ...',fontsize=15)

#设置x y轴label
plt.xlabel('x--lable')
plt.ylabel('y--lable')

#设置图例 显示
plt.legend(loc='up right')

#设置网格线
plt.grid(True)

#设置注解
plt.annotate(r'$\sqrt{2}$', (2, 1.3))

#显示绘图
plt.show()
```

### 直方图

- 直方图：一般对单维度的数据分析的时候使用，单维度的数据一般都是一个连续值

```python
x = np.random.randint(0,10,10)
plt.hist(x)
```

```python
#为什么说是概率呢
x
```

```python
#写一个简单的工资分布情况图
#定义一个列表，存储薪资的一些数据
salary = [2500, 3300, 2700, 5600, 6700, 5400, 3100, 3500, 7600, 7800, 
          8700, 9800, 10400]

group = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 11000]
plt.hist(salary, group, rwidth=0.8)
plt.legend()
plt.xlabel('salary-group')
plt.ylabel('salary')
plt.show()
```

```python
'''
构建不等距的直方图
'''
x = np.random.normal(100,20,100)
fig, ax = plt.subplots()
bins = [50, 60, 70, 90, 100,110, 140, 150]
ax.hist(x, bins, color="g",rwidth=0.8)
plt.show()
```

```python
#制作多类型的直方图
n_bins=10
fig,ax=plt.subplots(figsize=(8,5))
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax.hist(x_multi, n_bins, histtype='bar',label=list("ABC"))
ax.set_title('qiyue')
ax.plot()

plt.show()
```

### 饼图

饼图适合展示各部分占总体的比例，饼图只有一个参数x,使用pie()函数

- 普通各部分占满饼图
- 普通未占满饼图
- 饼图阴影、分裂等属性的设置

#### 占满饼图

```python
#普通各部分占满饼图
p = np.array([0.4,0.2,0.15,0.15,0.1])
plt.pie(p,labels=['dog','cat','brid','sheep','cow'],autopct='%0.1f',pctdistance=0.8)
```

```python
p = np.array([0.4,0.2,0.15])
plt.pie(p,labels=['dog','cat','brid'],shadow=True,explode=[0, 0.1, 0])
```

#### 未占满饼图

```python
p = np.array([0.4,0.2,0.1,0.1,0.1])
plt.pie(p,labels=['dog','cat','brid','sheep','cow'])
```

### 散点图

- 一般是挖掘关联关系的，探寻两个维度之间的关联关系
- 散点图是显示两组数据的值，每个点的坐标由变量的值决定
- 散点图最大用作用就是表明两个变量的相关性，相关性一般分为三种：正相关，负相关，不相关，

    由一组不连接的点完成，用于观察两种变量的相关性

例如，身高-体重，温度-纬度，等等

```python

#先创建x和y
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y)

#调整下他的颜色
plt.scatter(x,y,color='r',s=18)
```
